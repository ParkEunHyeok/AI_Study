```python
import cv2
image = cv2.imread('snoopy.png')

# 픽셀 수 및 이미지 크기 확인
print(image.shape)
print(image.size)

# 이미지 Numpy 객체의 특정 픽셀을 가리킵니다.
# 하나의 픽셀에 대한 정보
px = image[100, 100]

# B, G, R 순서로 출력됩니다.
# (단, Gray Scale인 경우에는 B, G, R로 구분되지 않습니다.)
print(px)

# R 값만 출력하기
print(px[2])
```

```python
# 특정 범위의 픽셀 변경

import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('snoopy.png')

# 반복문을 통한 픽셀 변경 연산
start_time = time.time()
for i in range(0, 100):
    for j in range(0, 100):
        image[i, j] = [255, 255, 255]
print("--- %s seconds ---" % (time.time() - start_time))

# 반복문보다 슬라이싱 연산을 통한 연산이 훨씬 빠름.
start_time = time.time()
image[0:100, 0:100] = [0, 0, 0]
print("--- %s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# ROI(관심있는 영역) 추출
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('snoopy.png')

# Numpy Slicing: ROI 처리 가능
# 추출한 크기와 붙여넣기하는 크기가 일치해야 함.
roi = image[140:240, 50:200]

# ROI 단위로 이미지 복사하기
image[0:100, 0:150] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 픽셀별 설정
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('snoopy.png')
image[:, :, 1] = 0

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
